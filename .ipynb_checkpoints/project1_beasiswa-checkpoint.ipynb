```python
# matrix X terlihat (y,x) (11,4) 11 baris, 4 kolom
X = np.array(df)
print(X)
```

```python
# menghitung rata rata IPK
ratarataIPK = np.mean(X[:,1])
print(ratarataIPK)
```

```python
# menghitung tingkat Kemiskinan dengan di kalikan 10
X[:,3] = X[:,3]*10
print(X[:,3])
```

```python
#membuat sebuah plot data
absis = X[:,1] # sumbu X (IPK)
ordinat = X[:,2] #sumbu Y (Tingkat Kemiskinan)

plt.scatter(absis,ordinat, label = "True Position")
plt.xlabel("IPK")
plt.ylabel("Tingkat Kemiskinan")
plt.title("Data Training")
plt.show()
```
